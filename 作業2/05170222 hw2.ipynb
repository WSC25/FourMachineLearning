```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('regression.csv')
x=data.iloc[:, 0].values
y=data.iloc[:, 1].values
```

```python
theta=[0,0]
eta=0.01
count=200
```

```python
# 定義計算梯度的函数
def compute_grad(theta, x, y):
    grad = [0, 0]
    grad[0] = 2. * np.mean(theta[0] + theta[1] * x - y)
    grad[1] = 2. * np.mean(x * (theta[0] + theta[1] * x - y))
    return np.array(grad)
```

```python
# 定義更新beta的函数
def update_beta(theta, eta, grad):
    new_beta = np.array(theta) - eta * grad
    return new_beta
```

```python
# 定義計算RMSE的函数
def rmse(theta, x, y):
    squared_err = (theta[0] + theta[1] * x - y) ** 2
    res = np.sqrt(np.mean(squared_err))
    return res
```

```python
# 進行第一次計算
grad = compute_grad(theta, x, y)
loss = rmse(theta, x, y)
theta = update_beta(theta, eta, grad)
loss_new = rmse(theta, x, y)
```

```python
# 開始迭代
i = 1
a = []
while i < 200:
    theta = update_beta(theta, eta, grad)
    grad = compute_grad(theta, x, y)
    loss = loss_new
    loss_new = rmse(theta, x, y)
    a.append(loss_new)
    i += 1
    print('Round %s Diff RMSE %s'%(i, abs(loss_new - loss)))
print('Coef: %s \nIntercept %s'%(theta[1], theta[0]))
```

```python
res = rmse(theta, x, y)
print('Our RMSE: %s'%res)
```
